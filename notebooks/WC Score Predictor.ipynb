```python
import pandas
import numpy
```

```python
def create_final_score(row):
    
    scoreline = str(row['score1']) + str(row['score2'])
    return '-'.join(sorted(scoreline, reverse=True))

def get_favourite(row):
    
    if row['our_score1'] > row['our_score2']:
        favourite = row['team1']
    elif row['our_score1'] < row['our_score2']:
        favourite = row['team2']
    else:
        favourite = 'Draw'
        
    return favourite

def is_correct_score(score1, score2, predicted_score1, predicted_score2):
    
    if score1 == predicted_score1 and score2 == predicted_score2:
        return True
    else:
        return False
    
def is_close_score(score1, score2, predicted_score1, predicted_score2):
    
    actual_score_difference = score2 - score1
    predicted_score_difference = predicted_score2 - predicted_score1
    
    if is_correct_result(score1, score2, predicted_score1, predicted_score2):
        if actual_score_difference == predicted_score_difference:
            if numpy.abs(score1 - predicted_score1) == 1 and numpy.abs(score2 - predicted_score2) == 1:
                return True
            else:
                return False
        elif numpy.abs(actual_score_difference - predicted_score_difference) == 1 and numpy.abs(score2 - predicted_score2) == 1 and score1 == predicted_score1:
            return True
        elif numpy.abs(actual_score_difference - predicted_score_difference) == 1 and numpy.abs(score1 - predicted_score1) == 1 and score2 == predicted_score2:
            return True
        else:
            return False

def is_correct_result(score1, score2, predicted_score1, predicted_score2):
    
    if score1 > score2 and predicted_score1 > predicted_score2:
        return True
    elif score1 < score2 and predicted_score1 < predicted_score2:
        return True
    elif score1 == score2 and predicted_score1 == predicted_score2:
        return True
    else:
        return False

def calculate_points_per_game(row):
    
    favourite = row['favourite']
    
    score1 = row['score1']
    score2 = row['score2']
    
    if row['team1'] == favourite:
        predicted_score1 = 2
        predicted_score2 = 0
    elif row['team2'] == favourite:
        predicted_score1 = 0
        predicted_score2 = 2
    else:
        predicted_score1 = 1
        predicted_score2 = 1
    
    if is_correct_score(score1, score2, predicted_score1, predicted_score2):
        points = 3
    elif is_close_score(score1, score2, predicted_score1, predicted_score2):
        points = 1.5
    elif is_correct_result(score1, score2, predicted_score1, predicted_score2):
        points = 1
    else:
        points = 0
        
    return points

def calculate_our_points_per_game(row):
    
    score1 = row['score1']
    score2 = row['score2']
    
    predicted_score1 = row['our_score1']
    predicted_score2 = row['our_score2']
    
    if is_correct_score(score1, score2, predicted_score1, predicted_score2):
        points = 3
    elif is_close_score(score1, score2, predicted_score1, predicted_score2):
        points = 1.5
    elif is_correct_result(score1, score2, predicted_score1, predicted_score2):
        points = 1
    else:
        points = 0
        
    return points
```

```python
data = pandas.read_csv('/Users/laurapallett/data/niclas_thomas/world_cup_results/results.csv')
data['final_score'] = data.apply(lambda row: create_final_score(row), axis=1)
data['favourite'] = data.apply(lambda row: get_favourite(row), axis=1)
```
